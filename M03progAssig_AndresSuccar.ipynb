```python
# 9.2

def get_odds():
    for number in range(10):
        if number % 2 != 0:
            yield number

# Using a loop to find and print the third odd number

count = 0
for odd in get_odds():
    count += 1
    if count == 3:
        print("Third odd number:", odd)
        break
```
